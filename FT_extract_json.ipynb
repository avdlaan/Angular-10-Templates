```python
data_dirs = ['/content/drive/MyDrive/train/document_parses/pdf_json'
            ]
json_article_paths = []
icnt = 0
data_list = []
for data_dir in data_dirs:
    json_article_paths = json_article_paths + glob.glob(os.path.join(data_dir, "*.json"))
    
    for json_path in json_article_paths:
      with open(str(json_path)) as f:
      #print(json_article_paths)
       data = json.load(f)
       text_str_2 = ""
       key = data.keys()
       for i in data['abstract']:
         text_str = str(i)
         print(text_str)
         data_list.append(text_str)
      f.close()
      
      icnt = icnt + 1
      if icnt == 30:
       break
print(len(json_article_paths))
```
